```python

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag
import re


nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

#  Input Paragraph
text = """The quick brown fox jumps over the lazy dog.
A smart student reads an interesting book."""

# Initialize NP List
NP_List = []

# STEP 5: Sentence Tokenization
sentences = sent_tokenize(text)
```

```python
# Process each sentence
for sentence in sentences:

    # Preprocessing (lowercase + clean)
    sentence = sentence.lower()
    sentence = re.sub(r'[^a-z\s]', '', sentence)

    # Word Tokenization
    words = word_tokenize(sentence)

    # POS Tagging
    tagged_words = pos_tag(words)

    # STEP 7: Define NP Grammar
    grammar = "NP: {<DT>?<JJ>*<NN|NNS|NNP|NNPS>+}"

    # Chunk Parsing
    chunk_parser = nltk.RegexpParser(grammar)
    tree = chunk_parser.parse(tagged_words)

    #  Extract NP
    for subtree in tree.subtrees(filter=lambda t: t.label() == 'NP'):
        np = " ".join(word for word, tag in subtree.leaves())
        NP_List.append(np)

print("Noun Phrases:\n")
for np in NP_List:
    print(np)
```

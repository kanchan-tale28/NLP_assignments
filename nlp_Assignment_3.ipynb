```python
import re
from collections import defaultdict

text_corpus = """Natural language processing is a field of artificial intelligence
that focuses on the interaction between computers and humans using natural language
processing techniques for better communication."""

N = 3
input_sequence = "natural language"
```

```python
# Preprocess Text
text = text_corpus.lower()
text = re.sub(r'[^a-z\s]', '', text)
words = text.split()


# Generate N-grams
ngram_dict = defaultdict(lambda: defaultdict(int))

for i in range(len(words) - N + 1):
    context = tuple(words[i:i+N-1])
    next_word = words[i+N-1]
    ngram_dict[context][next_word] += 1


# Calculate Probabilities
prob_dict = defaultdict(dict)

for context in ngram_dict:
    total_count = sum(ngram_dict[context].values())
    for next_word in ngram_dict[context]:
        prob_dict[context][next_word] = ngram_dict[context][next_word] / total_count
```

```python
# Predict Next Word
input_sequence = input_sequence.lower()
input_sequence = re.sub(r'[^a-z\s]', '', input_sequence)
input_words = input_sequence.split()

context = tuple(input_words[-(N-1):])

print("Given Context:", " ".join(context))

if context in prob_dict:
    predicted_word = max(prob_dict[context], key=prob_dict[context].get)
    print("Predicted Next Word:", predicted_word)
else:
    print("No prediction available")
```
